```python
def step_gradient(X,Y, learning_rate, m):
    M,N= X.shape
    m_slope = np.zeros(N)
    for i in range(M):
        x=X[i]
        y =Y[i]
        for j in range(N):
            m_slope[j] += (-2/M)* (y - (m * x).sum() )*x[j]
    new_m = m - learning_rate*m_slope
    return new_m
```

```python
def gd(X,Y, learning_rate, itt):
    N=X.shape[1]
    m=np.zeros(N)
    for i in range(itt):
        m=step_gradient(X,Y,learning_rate,m)
        print(i, " Cost: ", cost(X,Y, m))
    return m
```

```python
def cost(X,Y, m):
    total_cost = 0
    M = len(X)
    for i in range(M):
        x=X[i]
        y =Y[i]
        total_cost += (1/M)*((y - (m*x).sum())**2)
    return total_cost
```

```python
def run():
    learning_rate=0.1
    itt=600
    m = gd(X_train,Y_train, learning_rate, itt)
    return m
```

```python
m=run()
```

```python
def predict(X_test, m):
    M = X_train.shape[1]
    n_m = np.array(m).reshape(M, 1)
    return np.dot(X_test, n_m)
```
